# Minimal Demo Multilingual Impresso Named Entity Linking
This demo shows how to run the linking part on any text input.

```python
%pip install -q torch transformers requests requests-cache filelock
```

Initialize the pipeline that can process text with a marked named entity to resolve. The model is hosted on Hugging Face: https://huggingface.co/impresso-project/nel-mgenre-multilingual

```python
from transformers import AutoTokenizer, pipeline

NEL_MODEL_NAME = "impresso-project/nel-mgenre-multilingual"
nel_tokenizer = AutoTokenizer.from_pretrained(NEL_MODEL_NAME)

nel_pipeline = pipeline("generic-nel", model=NEL_MODEL_NAME,
                        tokenizer=nel_tokenizer,
                        trust_remote_code=True,
                        device='cpu')
```

Adapt the string by using the `[START]` and `[END]` to markup the entity.

```python
sentence1 = """
Le 0ctobre 1894, [START]Dreyfvs[END] est arrêté à Paris,
accusé d'espionnage pour l'Allemagne — un événement qui déch1ra
la société fr4nçaise pendant des années.
""".strip()

nel_pipeline(sentence1)
```
